# 숫자를 쉽게 다루는 numpy라이브러리

```python
import numpy as np
import random as rd

print(np.sqrt(2))

# <rd, np.rd 비교>

print(np.random.choice(6,10)) #0-5 사이 정수 랜덤하게 10개 추출
print(np.random.choice(6,10), p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]) # p값은 확률 그합은 1이어야함
print(rd.choice([1,3,4,5,6])) #리스트, 튜플, 문자열에서 하나 뽑아서 추출
#둘다 replace= False 로 복원추출을 막을 수 있다.

print(np.random.rand(5)) #0-1사이 실수 5개 추출
print(rd.randint(0,3)) #0-3 사이 정수 추출
```

# numpy이용해서 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

dice=np.random.choice(6,10000, p=[0.1,0.2,0.3,0.2,0.1,0.1]) # 뽑는 횟수 늘어날수록 확률에 가까워짐

plt.hist(dice, bins=6) # 각각 값의 횟수를 빈도수를 표시
plt.show()
```

# 버블차트그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.random.randint(10,100,200)
y= np.random.randint(10,100,200)
size= np.random.rand(200)*100

plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

# numpy array

```python
import numpy as np

a=np.array([1,2,3,4])
print(a)

b=np.array([1,3,'2',5])
print(b) #하나의 자료형만 저장 -> 정수가 문자로 변환

c=np.zeros(10)
print(c)

d=np.ones(10)
print(d)

e=np.eye(3)
print(e) #3 x 3 배열 생성
```

```python
import numpy as np

print(np.arange(3)) # 0 - 2 
print(np.arange(3,7)) # 3 - 6
print(np.arange(3,7,2)) # 3 - 6 까지 2간격으로

print(np.arange(1,2,0.1)) # 1 - 2 까지 0.1을 간격으로 출력
print(np.linspace(1,2,11)) # 1 -2 까지 11개 구간으로 나눈 실수 출력
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.zeros(10)+5
print(a)
b=np.ones(10)*5 #배열을 연산하면 한번에 모든 원소들이 연산된다.
print(b)

c=np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(c, np.sin(c))
plt.plot(c, np.cos(c))
plt.show()
```

  # mask

```python
import numpy as np

a=np.arange(-5,5)
mask1=a<0
print(mask1) #조건에 맞는 a의 요소만 true
print(a[mask1]) #a에 다시 mask를 적용하면 true인 원소만 추출

mask2=abs(a)>3 
print(a[mask1 + mask2]) # 두 mask의 합집합
print(a[mask1 * mask2]) # 두 mask의 교집합
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.randint(-100,100,1000)
y=np.random.randint(-100,100,1000)
size=np.random.rand(100)*100
mask1=abs(x)>50
mask2=abs(y)>50
#mask1=[true,false,false,true]
#mask2=[true,true,false,false]
x=x[mask1+mask2] #둘중에 하나라도 true인 x를 남김
y=y[mask1+mask2] #둘중에 하나라도 true인 y를 남김

plt.scatter(x,y,s=size,c=x ,cmap='jet')
plt.colorbar()
plt.show()
```
