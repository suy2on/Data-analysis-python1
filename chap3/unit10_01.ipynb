# 꺾은선 그래프로 표현하기 (각지점에서 비교하기가 편함)

```python
import csv

f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]

name=input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            num1=row[i].replace(',','')
            num2=row[i+103].replace(',','')
            m.append(int(num1))
            f.append(int(num2))
        break

import matplotlib.pyplot as plt

plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

# 막대그래프로 표현하기 (차이를 막대그래프로)

```python
import csv

f=open('gender.csv')
data=csv.reader(f)
d=[]

name=input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            num1=row[i].replace(',','')
            num2=row[i+103].replace(',','')
            d.append(int(num1)-int(num2))
        break

import matplotlib.pyplot as plt

plt.bar(range(101), d)
plt.rc('font',family='Malgun Gothic')
plt.title('남자인구와 여자인구의 차')
plt.show()
```

# 산점도로 데이터표현하기 (두변수사이의 관계파악)

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot') #스타일을 적용할 그래프 앞에 적어야함
plt.scatter([1,2,3,4],[10,30,20,40], s=[100,200,250,300], c=range(4), cmap='jet') 
plt.colorbar() #colorbar추가
#s는 사이즈, c는 색깔, cmap은 컬러바 색깔 정해줌
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x=[]
y=[]
size=[]

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.style.use('ggplot') #스타일을 적용할 그래프 앞에 적어야함
plt.scatter(x,y, s=size, c=size, cmap='jet', alpha=0.7) #alpha는 투명도 
plt.colorbar() #colorbar추가
#s는 사이즈, c는 색깔, cmap은 컬러바 색깔 정해줌
plt.show()
#여기서는 color가 size를 나타내줌
#c=x 라고 했다면 x축에 따라 색깔이 변했을 것
```

# 제주도의 연령대별 성별 비율을 산점도로 표현하기

```python
import csv

f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]

name=input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            num1=row[i].replace(',','')
            num2=row[i+103].replace(',','')
            m.append(int(num1))
            f.append(int(num2))
        break

import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.scatter(m,f, c=range(101), cmap='jet', alpha=0.7)# color가 연령을 나타낼수있도록 range(101)
plt.colorbar()
plt.rc('font', family='Malgun Gothic')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.plot(range(max(m)),range(max(m)),'g') # y=x추세선 추가
plt.show()
```
