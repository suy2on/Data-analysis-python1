# 제주도 남여 성비 알아보기 (bar그래프)

```python
import csv

f=open('gender.csv')
data=csv.reader(f)
m=[] #남성 0세부터 100세이상까지 (총101개)리스트에 넣기
f=[] #여성 0세부터 100세이상까지 (총101개)리스트에 넣기

for row in data:
    if '제주'in row[0]:
        for i in range(101):
            num1=row[i+3].replace(',', '') #2,139 을 2139로 바꿔주기
            num2=row[-(i+1)].replace(',', '')
            m.append(-int(num1)) #두개를 한번에 나타낼때 항아리 모양으로 만들어주기위해 한 데이터를 음수로 바꿔준다
            f.append(int(num2))
        break
            
f.reverse() #여자 리스트를 역순으로 돌려준다

#import matplotlib 

#matplotlib.font_manager._rebuild() #폰트 불러오지 못하고 warning뜰때 rebuild해주기

import matplotlib.pyplot as plt

plt.style.use('ggplot') #격자모양
plt.rc('font', family='Malgun Gothic') #title에 한글 쓸 때 짝꿍
plt.rcParams['axes.unicode_minus']=False #-부호 깨지지 않게
plt.title('제주특별자치도 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m, label='남성')
plt.barh(range(101),f, label='여성')
plt.legend() #label과 짝꿍
plt.show()
```

# 혈액형 pie그래프로 나타내기

```python
#비율을 나타낼 때 좋다

import matplotlib.pyplot as plt

size=[2441, 2312, 1031, 1233]
label=['A형','B형','AB형','O형']
color=['darkmagenta','deeppink','hotpink','pink'] #색깔
plt.axis('equal') # 동그란 파이모형으로 만들어주기
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.2,0)) #labels로 라벨달아주기 
#autopct로 소수점 첫째자리까지 퍼센트 비율을 나타내줌
#explode는 각각의 돌출정도를 표시함 0은 돌출 안된것 
plt.legend()
plt.show() #pie의 시작은 3시방향에서 시계반대방향
```

# 제주도의 성별 인구 비율 표현하기

```python
import csv

f=open('gender.csv')
data=csv.reader(f)
m=[] #남성 0세부터 100세이상까지 (총101개)리스트에 넣기
f=[] #여성 0세부터 100세이상까지 (총101개)리스트에 넣기

for row in data:
    if '제주'in row[0]:
        for i in range(101):
            num1=row[i+3].replace(',', '') #2,139 을 2139로 바꿔주기
            num2=row[-(i+1)].replace(',', '')
            m.append(int(num1)) #두개를 한번에 나타낼때 항아리 모양으로 만들어주기위해 한 데이터를 음수로 바꿔준다
            f.append(int(num2))
        break
            
f.reverse() #여자 리스트를 역순으로 돌려준다

import matplotlib.pyplot as plt

label=['남성','여성']
plt.axis('equal')
plt.pie([sum(m),sum(f)], autopct='%.1f%%', labels=label, startangle=90) #startangle은 시작각도조정
plt.legend()
plt.rc('font', family='Malgun Gothic')
plt.title('제주특별자치구 지역의 남녀 성별 비율')
plt.show()
```
