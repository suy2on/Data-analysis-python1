# 내 동네 인구 구조 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data=csv.reader(f)
next(data)
age=[]

for row in data:
    #if row[0]=='서울특별시 마포구 서교동(1144066000)':
    if '서교동' in row[0]: # 서교동이라는 글이 row[0]안에 있기만 하면 true
        for i in range(101):
            age.append(int(row[i+3]))
        #for i in row[3:]:   
            #age.append(int(i))
print(len(age)) #101개가 출력됐는지 확인
plt.style.use('ggplot') #격자무늬로 변경
plt.plot(age)
plt.show()
               
               
```

# 원하는 동네 인구분포표 뽑기

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data=csv.reader(f)
next(data)
age=[]

town=input('인구 구조를 알고 싶은 지역의 이름을 입력해 주세요:')

for row in data:
    if town in row[0]: 
        for i in row[3:]:   
            age.append(int(i))
            
plt.style.use('ggplot') #격자무늬로 변경
plt.rc('font', family='Malgun Gothic') #한글입력하기위해
plt.title(town+'의 인구 구조') 
plt.plot(age)
plt.show()
```
