# 막대 그래프그리기

```python
import matplotlib.pyplot as plt

plt.bar([0,3,2,1,6,7], [1,2,3,5,6,7]) #막대표시할위치, 막대높이
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('age.csv')
data=csv.reader(f)
next(data)
age=[]

for row in data:
    #if row[0]=='서울특별시 마포구 서교동(1144066000)':
    if '서교동' in row[0]: # 서교동이라는 글이 row[0]안에 있기만 하면 true
        for i in range(101):
            age.append(int(row[i+3]))
        #for i in row[3:]:   
            #age.append(int(i))


#plt.bar(range(101),age)  수직으로 그리기
plt.barh(range(101), age) # 수평으로 그리기
plt.show()
```

# 항아리 그래프 그리기

```python
import csv

f=open('gender.csv')
data=csv.reader(f)
m=[] #남성 0세부터 100세이상까지 (총101개)리스트에 넣기
f=[] #여성 0세부터 100세이상까지 (총101개)리스트에 넣기

for row in data:
    if '서교'in row[0]:
        for i in range(101):
            m.append(-int(row[i+3])) #두개를 한번에 나타낼때 항아리 모양으로 만들어주기위해 한 데이터를 음수로 바꿔준다
            f.append(int(row[-(i+1)]))
            
f.reverse() #여자 리스트를 역순으로 돌려준다

import matplotlib 
matplotlib.font_manager._rebuild() #폰트 불러오지 못하고 warning뜰때 rebuild해주기

import matplotlib.pyplot as plt

plt.style.use('ggplot') #격자모양
plt.rc('font', family='Malgun Gothic') #title에 한글 쓸 때 짝꿍
plt.rcParams['axes.unicode_minus']=False #-부호 깨지지 않게
plt.title('서교동 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m, label='남성')
plt.barh(range(101),f, label='여성')
plt.legend() #label과 짝꿍
plt.show()
```
