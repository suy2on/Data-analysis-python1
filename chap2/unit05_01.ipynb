# 1907년부터 최고온도 리스트로 뽑기

```python
import csv

f=open('..\chap1\seoul.csv') #..는 상위 폴더 거기에서 chap1에 있는 seoul.csv
data=csv.reader(f)
next(data)  #첫줄날리기
result=[]

for row in data:
    if row[-1]!='':
        result.append(float(row[-1]))

print(result)    
print(len(result))
```

# 8월간 최대온도 그래프

```python
import csv
import matplotlib.pyplot as plt

f=open('..\chap1\seoul.csv') #..는 상위 폴더 거기에서 chap1에 있는 seoul.csv
data=csv.reader(f)
next(data)  #첫줄날리기
result=[]

for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='08': #row[0]인 날짜부분을 split해서 [년,월,일] 그중 8월들만 뽑음
            result.append(float(row[-1]))


plt.plot(result, "hotpink")
f.close()

```

# 1983년 이후 내 생일의 최저, 최고온도 그래프

```python
f=open('..\chap1\seoul.csv') #..는 상위 폴더 거기에서 chap1에 있는 seoul.csv
data=csv.reader(f)
next(data)  #첫줄날리기
high=[]
low=[]
year=[]

for row in data:
    if row[-1]!='' and row[-2]!='':
        if 1983<=int(row[0].split('-')[0]): #1983년 이후로
            if row[0].split('-')[1]=='10'and row[0].split('-')[2]=='17': #row[0]인 날짜부분을 split해서 [년,월,일] 그중 8월들만 뽑음
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                year.append(float(row[0].split('-')[0]))


plt.plot(year,high, "r", label='high')
plt.plot(year,low, 'b', label='row')
plt.rc('font', family='Malgun Gothic')  #한글 안깨지게 하는 코드
plt.rcParams['axes.unicode_minus']=False # -값 안깨지게 하기
plt.title("내 새일의 기온 변화 그래프")
plt.legend(loc=2)
plt.show()
f.close()
```
