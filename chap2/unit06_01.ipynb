# 기온 데이터를 다양하게 시각화하기

# 히스토그램

```python
import matplotlib.pyplot as plt
import random as rd

dice=[]
for i in range(1000000): #반복하는 수가 클수록 균등해진다!!
    dice.append(rd.randint(1,6))

plt.hist(dice, bins=6) #bins는 가로축의 구간 개수

```

# 기온 데이터를 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('..\chap1\seoul.csv')
data= csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1]!='':
        result.append(float(row[-1]))
                     
plt.hist(result, bins=100, color='red')
plt.show()                      

                      
                      
```

```python
f=open('..\chap1\seoul.csv')
data= csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='08':
            result.append(float(row[-1]))
                     
plt.hist(result, bins=100, color='red')
plt.show()  #8월의 날씨만 모으면 정규분포
```

# 기온 데이터를 상자 그림으로 표현하기(boxplot)

```python
import matplotlib.pyplot as plt
import random
import numpy as np

result=[]
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))

plt.boxplot(result)# 최소값, 1/4값, 2/4값, 3/4값 최대값을 차례로 나타내주는 그림
plt.show()

result=np.array(result)
print("1/4:" +str(np.percentile(result,25))) #1/4위치에 해당하는 값 뽑기
```

# 월별 최고기온 그래프

```python
import csv
import matplotlib.pyplot as plt

f=open('..\chap1\seoul.csv')
data= csv.reader(f)
next(data)
month=[[],[],[],[],[],[],[],[],[],[],[],[]] #리스트안에 리스트 초기화 해주기

for row in data:
    if row[-1]!='':
            month[int(row[0].split('-')[1])-1].append(float(row[-1]))
                     
plt.boxplot(month, showfliers=False) # showfliers=False 는 이상치를 없애는 것
plt.show()                      
```
