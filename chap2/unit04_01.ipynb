# 기본 그래프 그리기

```python
import matplotlib.pyplot as plt

plt.title('plotting')
plt.plot([10,20,30,40], 'r^--', label='asc')
#plt에 있는 plot 함수: 리스트 값은 y값 , x값은 자동으로 0부터 1씩증가
#label은 그래프의 이름을 붙여주는 것 legned와 같이씀
#r^-- : 색상,마커모양,선모양 한번에 설정 그대신 위치중요!! 좌표값 바로뒤에 위치

plt.plot([1,2,3,4],[12,43,25,15], label='curve', ls=':', marker='.')
#첫번째 데이터는 x, 두번째 데이터는 y   
#ls= linestyle의 약자 그대로 써도 된다
plt.legend() #두개이상의 데이터의 범례를 나타내줌 
plt.show
```
