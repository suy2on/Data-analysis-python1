# 출근 시간대 사람들이 가장 많이 타고 내리는 역

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)
result=[]

for row in data:
    row[4:]=map(int, row[4:]) # 반복문으로 할 일을 map(함수, 적용할 데이터)함수로 
    result.append(sum(row[10:15:2])) # 인덱스 10,12,14를 더해서 append

result.sort()
print(result)
plt.bar(range(len(result)), result)
plt.show()
```

# 몇시에 사람들이 가장 많이 타는 역은 어디일까?

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)
time=int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? : ')) #문자열->정수형
max=0
station=''

for row in data:
    row[4:]=map(int, row[4:]) # 반복문으로 할 일을 map(함수, 적용할 데이터)함수로 
    if max< row[4+(time-4)*2]: 
        max=row[4+(time-4)*2]
        station=row[3]+'('+row[1]+')'
print(station, max)
```

# 시간대별로 사람들이 가장 많이 타고 내리는 역

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

max=[0]*24
station=['']*24

for row in data:
    row[4:]=map(int, row[4:]) # 반복문으로 할 일을 map(함수, 적용할 데이터)함수로 
    for j in range(24):
        i=j*2+4 # 4,6,8....
        if max[j]< row[i]: 
            max[j]=row[i]
            station[j]=row[3]+'('+str(j+4)+'시)'
print(max)
print(station)

plt.rc('font', family='Malgun Gothic')
plt.title('시간대별 최대승차역')
plt.bar(range(24),max)
plt.xticks(range(24), station,rotation=90) # x값에 이름넣기
plt.show()
```

# 모든 지하철에 승하차 인원 시간대 별로 보기

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

on=[0]*24
out=[0]*24

for row in data:
    row[4:]=map(int, row[4:]) # 반복문으로 할 일을 map(함수, 적용할 데이터)함수로 
    for j in range(24):
        i=j*2+4 # 승차 4,6,8...  
        k=j*2+5 # 하차 5,7,9...
        on[j]=on[j]+row[i] # 승차 합
        out[j]=out[j]+row[k]# 하차 합

plt.rc('font', family='Malgun Gothic')
plt.title('시간대별 지하철 승하차 인원')        
plt.plot( on, label='승차')
plt.plot( out, label='하차')
plt.xticks(range(24), range(4,28)) # x좌표값을 4-27까지 나타내는 
plt.legend()
plt.show() #그래프 y축 위에 1e7은 1*10의 7승 즉 천만명이 단위
  
```
